```python
#Histogram
import matplotlib.pyplot as plt
plt.hist(df["Age"], bins = 6)
```

```python
import numpy as np
cgpa = [6, 6.8, 7, 7.5, 8, 8.5, 10]
package = [5, 6, 7, 8, 9, 12, 15]
cgpa = np.array(cgpa)
package = np.array(package)
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}
plt.title("Data of shoolini university", fontdict = font1)
plt.xlabel("cgpa", fontdict = font2)
plt.ylabel("package", fontdict = font2)

plt.plot(cgpa, package, "*:g", ms = 20, mec = "r", mfc = "b")
plt.grid()
plt.show()
```

```python
fruit = np.array(["apple", "Banana", "Mango", "Orange"])
fruit_number = np.array([40, 50, 30, 100])
plt.title("Fruits data")
plt.xlabel("Fruits")
plt.ylabel("Fruits number")
color = ["red", "yellow", "pink", "orange"]
plt.bar(fruit, fruit_number, color = color)
plt.show()
```
